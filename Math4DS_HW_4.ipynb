```python
import numpy as np
from scipy.misc import derivative 
import matplotlib.pyplot as plt
```

### Уровень 0:


Посчитайте значение первой и второй производной функции:

$f(x) = x^5 + 4sin(2x) + cos(3x+3)$ 

В точке x = 1.



#### Вариант математика :)

```python
x = 1
func = x**5 + 4*np.sin(2*x) + np.cos(3*x + 3)
dev_1 = 5*x**4 - 3*np.sin(3*x + 3) + 8*np.cos(2*x)
dev_2 = 20*x**3 - 16*np.sin(2*x) - 9*np.cos(3*x + 3)
print(f"В точке х=1, \nфункция равна: {func} \nпервая производная равна: {dev_1} \nвторая производная равна: {dev_2}")
```

#### Вариант программиста :)

```python
def f(x):
    return x**5 + 4 * np.sin(2*x) + np.cos(3*x + 3)

func = f(1)
dev_1 = derivative(f, 1.0, dx=1e-6)
dev_2 = derivative(f, 1.0, dx=1e-6, n=2)

print(f"В точке х=1, \nфункция равна: {func} \nпервая производная равна: {dev_1} \nвторая производная равна: {dev_2}")
```

### Уровень 1:


Постройте график первой и второй производной функции:

$f(x) = sin^5(2x+1)$

На отрезке [-5; 5].

```python
def f(x):
    return (np.sin(2 * x + 1)) ** 5

interval = np.linspace(-5, 5)

func = f(interval)
dev_1 = [derivative(f, x, dx=1e-6, n=1) for x in interval]
dev_2 = [derivative(f, x, dx=1e-6, n=2) for x in interval]

fig, ax = plt.subplots()

ax.plot(interval, func,   
         linestyle = '--', 
         linewidth = 4, 
         color = 'darkmagenta')

ax.plot(interval, dev_1,  
         linestyle = '-.',
         linewidth = 3,
         color = 'indigo')

ax.plot(interval, dev_2,  
         linestyle = ':',
         linewidth = 2,
         color = 'darkblue')

plt.title("График  функции, первой и второй производной")
ax.legend(["Функция", "1-я производная", "2-я производная"])

fig.set_figwidth(12)
fig.set_figheight(6)
fig.set_facecolor('linen')
ax.set_facecolor('ivory')

plt.show()
```
