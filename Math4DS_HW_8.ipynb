### Известно, что светофор горит 54 секунды красным, 4 секунды жёлтым и 28 секунд зелёным. Посчитайте, какое количество информации несёт сообщение о цвете светофора в текущий момент. 

Количество информации можно посчитать по формуле Шэннона:
$H(x) = -\sum\limits_{i=1}^n p_i \cdot log_2 p_i$  
Вероятность для каждого цвета посчитаем как отношение продолжительности его горения к продолжительности одного цикла светофора.

```python
red = 54
yellow = 4
green = 28
itog = red + yellow + green


inf = - (red/itog * math.log2(red/itog) + yellow/itog * math.log2(yellow/itog) + green/itog * math.log2(green/itog))
print(f'Ответ {inf}.\nСообщение о цвете светофора в текущий момент несет в себе количество информации, равное {math.ceil(inf)} бит.')
```
