### Скорость метода loc
Посчитаем средний рейтинг разных жанров. Метод merge взят из материалов следующего занятия ¯ \ _ (ツ) _ / ¯

````python
ratings = pd.read_csv('ml-latest-small/ratings.csv')
movies = pd.read_csv('ml-latest-small/movies.csv')
joined = ratings.merge(movies, how='left', on='movieId')
joined.head()
````

````python
joined['Adventure'] = joined.apply(lambda row: row.rating if 'Adventure' in row.genres else None, axis=1)
joined.head()
````

Можно и через loc

````python
joined['Adventure'] = joined.loc[joined.genres.str.contains('Adventure'), 'rating']
joined.head()
````

````python
joined['Adventure'] = joined.loc[joined.genres.str.contains('Adventure'), 'title']
joined.head()
````

Для замера времени сделаем эти операции для набор жанров

````python
genres = ['Adventure', 'Animation', 'Children', 'Drama', 'Musical', 'Thriller']
````

````python
%%time
joined = ratings.merge(movies, how='left', on='movieId')

for genre in genres:
    joined[genre] = joined.apply(lambda row: row.rating if genre in row.genres else None, axis=1)
````

````python
%%time
joined = ratings.merge(movies, how='left', on='movieId')

for genre in genres:
    joined[genre] = joined.loc[joined.genres.str.contains(genre), 'rating']
````

**Задание 1**
Для датафрейма log из материалов занятия создайте столбец source_type по следующим правилам:
- если источник traffic_source равен yandex или google, то в source_type ставится organic
- для источников paid и email из России - ставим ad
- для источников paid и email не из России - ставим other
- все остальные варианты берем из traffic_source без изменений

````python
log = pd.read_csv('visit_log.csv', sep=';')
log.head()
````

````python
# Вариант 1 - через apply

def get_source_type(query):
    if query['traffic_source'] == 'yandex' or query['traffic_source'] == 'google':
        source_type = 'organic'
    elif query['traffic_source'] == 'paid' or query['traffic_source'] == 'email':
        if query['region'] == 'Russia':
            source_type = 'ad'
        else:
            source_type = 'other'
    else:
        source_type = query['traffic_source']
    return source_type
````

````python
%%time
log['source_type'] = log.apply(get_source_type, axis = 1)
````

````python
log.head(10)
````

````python
log.traffic_source.unique()
````

````python
%%time
# Вариант 2 - через loc, другое название столбца чтобы сравнить оба варианта

russia_dict = {'Russia' : {'yandex': 'organic', 'google': 'organic', 'paid': 'ad', 'email': 'ad', 'direct': 'direct'}, 'Other' : {'yandex': 'organic', 'google': 'organic', 'paid': 'other', 'email': 'other', 'direct': 'direct'}}
for key in russia_dict.keys():
    if key == 'Russia':
        for key, value in russia_dict['Russia'].items():
            log.loc[(log.region == 'Russia') & (log.traffic_source == key), 'traffic_type'] = value
    if key != 'Russia':
        for key, value in russia_dict['Other'].items():
            log.loc[(log.region != 'Russia') & (log.traffic_source == key), 'traffic_type'] = value
log.head(10)
````
